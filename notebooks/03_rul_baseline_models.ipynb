# 03 RUL Baseline Models

## Purpose
Train and evaluate interpretable RUL baselines as the engineering reference point.

## Outputs
- Baseline metrics (MAE/RMSE)
- Error diagnostics for reliability interpretation
- Saved artifacts under results/metrics and results/tables

## Assumptions
- RUL targets follow standard CMAPSS construction
- Feature generation is leakage-safe and based on prior/current cycles only
- Comparisons prioritize interpretability before sequence models
