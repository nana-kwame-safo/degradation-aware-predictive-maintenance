# 02 Preprocessing and Splits

## Purpose
Define leakage-safe preprocessing decisions and unit-based train/validation splits.

## Outputs
- Validated unit-based split configuration
- Train-fit scaling approach for sensor columns
- Windowing parameters for downstream models

## Assumptions
- Splits are performed by unit_id, never by row
- Scaling is fit on train split only
- Configuration matches src/config.py defaults unless explicitly documented
